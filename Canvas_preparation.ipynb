Importing the dataset and libraries.

Renaming the dataset for better understanding.
This dataset relates to if and when (the location will be given when we cross-reference the ID with the other datasets) a canvas was placed on the location: so in_place it means it was placed, justification means why it was placed, length the length of the canvas, and num_point the number of points it was placed at that specific location.

```python
canvas = canvas.rename(columns={'ano':'year','mes':'month','processo_numero':'ID','colocacao_lona_situacao':'in_place',
                      'colocacao_lona_data':'date_of_placement','colocacao_lona_justificativa':'justification',
                      'colocacao_lona_metragem':'length','colocacao_lona_qtd_pontos':'num_points'})
canvas
```

Isolating the month and date values in the respective columns.

```python
aux=[]
for i in range (0,len(canvas['month'])):
    aux = canvas['month'][i]
    canvas['month'][i] = aux[:2]
    if canvas['month'][i] == 'Nã':
        canvas['month'][i] = 0
    canvas['month'][i] = int(canvas['month'][i])

canvas['month'][:10]
```

```python
canvas.head()
```

```python
canvas['date_of_placement'].fillna(0,inplace=True)
aux=[]
#aux = canvas['date_of_placement'][4]
print(aux[:2])
for i in range (0,len(canvas['date_of_placement'])):
    if canvas['date_of_placement'][i] !=0:
        aux = canvas['date_of_placement'][i]
        canvas['date_of_placement'][i] = aux[:2]
        canvas['date_of_placement'][i] = int(canvas['date_of_placement'][i])

canvas['date_of_placement'][:10]
```

```python
canvas.head()
```

Transforming the in_place column into numerical values.

```python
canvas = pd.get_dummies(canvas, columns=['in_place'],drop_first=True)
```

```python
canvas.head()
```

```python
canvas = canvas.rename(columns={'in_place_sim':'in_place'})
```

After a look at the justification column, we detected that the information that we need is if a canvas was placed due to a sliding, so we'll isolate that keyword 'deslizamento'(sliding in portuguese) and transform into a 0, 1 column respectively.

```python
import re
canvas['justification'] = canvas['justification'].astype(str)

text = []
for i in range(0, len(canvas['justification'])):
    review = re.sub('[^a-zA-Z]', ' ', canvas['justification'][i])
    review = review.lower()
    review = review.split()
    text.append(review)


            

canvas['justification'] = text
```

```python
canvas.head()
```

```python
canvas['justification'] = [1 if 'deslizamento' in justification else 0 for justification in canvas['justification']]
canvas.head()
```

Finally there are some duplicate ID entries, so we'll drop the ones that might have 0 values in the event of a sliding.

```python
canvas = canvas.sort_values('justification', ascending=False).drop_duplicates('ID').sort_index()
```

```python
canvas.to_csv(path_or_buf='canvas_prepared.csv')
```
