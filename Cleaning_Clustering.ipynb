```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


solicita=pd.read_csv("sedec_solicitacoes.csv",delimiter =';')
chamados=pd.read_csv("sedec_chamados.csv",delimiter =';')
lonas=pd.read_csv("sedec_lonas.csv",delimiter =';')
ocorrencia=pd.read_csv("sedec_tipo_ocorrencias.csv",delimiter =';')
vistoria=pd.read_csv("sedec_vistorias.csv",delimiter =';')
```

```python
chamados = chamados.rename(columns={'processo_numero':'ID'})
solicita = solicita.rename(columns={'processo_numero':'ID'})
lonas = lonas.rename(columns={'processo_numero':'ID'})
ocorrencia = ocorrencia.rename(columns={'processo_numero':'ID'})
vistoria = vistoria.rename(columns={'processo_numero':'ID'})

chamados.drop(columns='processo_origem', axis=1, inplace=True)
solicita.drop(columns='processo_origem', axis=1, inplace=True)
```

```python
uniao = solicita.set_index("ID").join(chamados.set_index('ID'))
```

```python
uniao['processo_situacao'] = uniao['processo_situacao'].astype('str')
uniao['solicitacao_descricao'] = uniao['solicitacao_descricao'].astype('str')
uniao['solicitacao_vitimas'] = uniao['solicitacao_vitimas'].astype('str')
uniao['solicitacao_vitimas_fatais'] = uniao['solicitacao_vitimas_fatais'].astype('str')
uniao['solicitacao_descricao'].str.lower()
```

```python
uniao['solicitacao_descricao']=[1 if 'deslizamento' in review else 0 for review in uniao['solicitacao_descricao']]
uniao['processo_situacao']=[1 if sit == 'completo' else 0 for sit in uniao['processo_situacao']]
uniao['solicitacao_vitimas']=[1 if sit == 'Sim' else 0 for sit in uniao['solicitacao_vitimas']]
uniao['solicitacao_vitimas_fatais']=[1 if sit == 'Sim' else 0 for sit in uniao['solicitacao_vitimas_fatais']]
```
