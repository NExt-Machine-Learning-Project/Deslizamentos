```python
import pandas as pd
import numpy as np


callings=pd.read_csv("Data\sedec_chamados.csv",delimiter =';')
```

```python
callings = callings.rename(columns={'processo_numero':'ID'})



callings.drop(columns='processo_origem', axis=1, inplace=True)
```

```python
callings.info()
```

```python
callings.head()
```

```python
callings.drop(columns='ano', axis=1, inplace=True)
callings.drop(columns='mes', axis=1, inplace=True)
callings.drop(columns='solicitacao_data', axis=1, inplace=True)
callings.drop(columns='solicitacao_descricao', axis=1, inplace=True)
callings.drop(columns='solicitacao_regional', axis=1, inplace=True)
callings.drop(columns='solicitacao_bairro', axis=1, inplace=True)
callings.drop(columns='solicitacao_localidade', axis=1, inplace=True)
callings.drop(columns='solicitacao_endereco', axis=1, inplace=True)
callings.drop(columns='solicitacao_roteiro', axis=1, inplace=True)
callings.drop(columns='rpa_codigo', axis=1, inplace=True)
callings.drop(columns='rpa_nome', axis=1, inplace=True)
callings.drop(columns='solicitacao_microrregiao', axis=1, inplace=True)
callings.drop(columns='solicitacao_plantao', axis=1, inplace=True)
callings.drop(columns='processo_tipo', axis=1, inplace=True)
callings.drop(columns='processo_situacao', axis=1, inplace=True)
callings.drop(columns='solicitacao_hora', axis=1, inplace=True)
callings.head()
```

```python
callings['solicitacao_vitimas'] = callings['solicitacao_vitimas'].astype('str')
callings['solicitacao_vitimas_fatais'] = callings['solicitacao_vitimas_fatais'].astype('str')
```

```python
callings['solicitacao_vitimas']=[1 if sit == 'Sim' else 0 for sit in callings['solicitacao_vitimas']]
callings['solicitacao_vitimas_fatais']=[1 if sit == 'Sim' else 0 for sit in callings['solicitacao_vitimas_fatais']]
```

```python
callings.head()
```

```python
callings.drop(columns='solicitacao_origem_chamado', axis=1, inplace=True)
callings.drop(columns='processo_status', axis=1, inplace=True)
callings.drop(columns='processo_localizacao', axis=1, inplace=True)
callings.drop(columns='processo_data_conclusao', axis=1, inplace=True)
callings.head()
```

```python
callings = callings.rename(columns={'solicitacao_vitimas':'victims',
                                 'solicitacao_vitimas_fatais':'deadly_victims'})
callings.head()
```

```python
callings.info()
```

```python
callings = callings.drop_duplicates('ID')
```

```python
callings.info()
```

```python
callings.fillna(0,inplace=True)
callings.head()
```

```python
callings['latitude'].value_counts()
```

```python
callings.to_csv(path_or_buf='callings_prepared.csv')
```
