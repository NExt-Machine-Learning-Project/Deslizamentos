```python
lonas = lonas.rename(columns={'ano':'year','mes':'month','processo_numero':'ID','colocacao_lona_situacao':'in_place',
                      'colocacao_lona_data':'date_of_placement','colocacao_lona_justificativa':'justification',
                      'colocacao_lona_metragem':'length','colocacao_lona_qtd_pontos':'num_points'})
lonas
```

```python
aux=[]
for i in range (0,len(lonas['month'])):
    aux = lonas['month'][i]
    lonas['month'][i] = aux[:2]
    if lonas['month'][i] == 'Nã':
        lonas['month'][i] = 0
    lonas['month'][i] = int(lonas['month'][i])

lonas['month'][:10]
```

```python
lonas.head()
```

```python
lonas['date_of_placement'].fillna(0,inplace=True)
aux=[]
#aux = lonas['date_of_placement'][4]
print(aux[:2])
for i in range (0,len(lonas['date_of_placement'])):
    if lonas['date_of_placement'][i] !=0:
        aux = lonas['date_of_placement'][i]
        lonas['date_of_placement'][i] = aux[:2]
        lonas['date_of_placement'][i] = int(lonas['date_of_placement'][i])

lonas['date_of_placement'][:10]
```

```python
lonas.head()
```

```python
lonas = pd.get_dummies(lonas, columns=['in_place'],drop_first=True)
```

```python
lonas.head()
```

```python
import re
lonas['justification'] = lonas['justification'].astype(str)

text = []
for i in range(0, len(lonas['justification'])):
    review = re.sub('[^a-zA-Z]', ' ', lonas['justification'][i])
    review = review.lower()
    review = review.split()
    text.append(review)


            

lonas['justification'] = text
```
